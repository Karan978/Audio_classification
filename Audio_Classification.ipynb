```python
y=[]
for i in spec:
    a=i.replace('.png','.wav')
    b = int(d[a])
    y.append(b)
    
```

```python
lb = LabelEncoder()

y = np_utils.to_categorical(lb.fit_transform(y))
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
type(X_train)
```

```python
def classification_model():
    ip = Input((217,223,3))
    
    conv1 = Conv2D(12,(5,5),padding='same',activation='relu')(ip)
    
    conv2 = Conv2D(24,(5,5),padding='same',activation='relu')(conv1)
    
    conv3 = Conv2D(36,(3,3),padding='same',activation='relu')(conv2)
    
    conv4 = Conv2D(48,(3,3),padding='same',activation='relu')(conv3)
    
    pool = MaxPool2D(2,2)(conv4)
    
    drop = Dropout(0.25)(pool)
    
    flatten = Flatten()(drop)
    
    dense1 = Dense(40,activation='relu')(flatten)
    
    dense2 = Dense(20,activation='relu')(dense1)
    
    dense3 = Dense(10,activation='softmax')(dense2)
    
    model = Model(inputs = ip,outputs = dense3)
    
    return model
    
```

```python
model = classification_model()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
  
```

```python
callbacks = [
    EarlyStopping(patience=10, verbose=1),
    ModelCheckpoint('model-audio-classification.h5', verbose=1, save_best_only=True, save_weights_only=True)
]
```

```python
results = results = model.fit(X_train, y_train, batch_size=16,validation_data=(X_valid, y_valid), epochs=50, callbacks=callbacks)
```
